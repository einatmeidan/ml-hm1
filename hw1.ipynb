## 8. Mini-Batch Gradient Descent (10 pts)

In **mini-batch gradient descent** we update the parameters using the gradient computed on a *small batch* of training samples at a time, instead of using the full dataset (full‑batch GD). This is more computationally efficient, but leads to updates based on an approximation to the true gradient. Nevertheless, in many cases this approximation is good enough to attain fast convergence to the minimum loss. 

The size of the batches used in each update is fixed to some small value (for example. 32 or 64). 
Training is partitioned into **epochs**, where in each epoch the entire training dataset is traversed once.
In each epoch, the dataset is randomly partitioned into batches of the predetermined size.
Each epoch then proceeds with a series of update steps,
and in each step the model parameters are updated according to an approximation of the gradient that is based on a different batch of samples. In this version we do not use an early stop condition as we had with standard GD.

**Remark:** There is a slightly different variant of mini-batch gradient descent, in which a random batch of the pre-fixed size is selected in each update. This version does not have epochs and is often referred to as **mini-batch stochastic gradient descent (SGD)** or simply **stochastic gradient descent (SGD)**. You will implement the more standard **mini-batch gradient descent** described above.

## 9. Regression with Regularization (15 pts)

In Question 2 of the theretical execise in this assignment, we considered L1-regularized least squared linear regression (LASSO). Recall that the loss function in this version of regularized linear regression is:
$$
\bar{J}_\lambda(w; D) 
~~\triangleq~~ \frac 1n\sum_{i=1}^n  \left( w^\top x^{(i)} - y^{(i)} \right)^2 + \lambda \sum_{j=0}^p |w_j|
~~=~~ \frac 1n \left\|Xw-y\right\|^2 + \lambda\left\|w\right\|_1 ~.
$$
Intuitively, the regularization term, $\lambda \sum_{j=0}^p |w_j|$, encourages solutions with **smaller weight norms**, which can help when features are correlated, when the design matrix $X$ is singular, or when we want to avoid overfitting to noise in the training data. The hyperparameter $\lambda \ge 0$ controls the trade‑off between fitting the data (small squared error) and keeping the weights small.
Note that for the squared error term of the loss we use the **mean** squared error (MSE), to maintain consistency with previous analyses.

In this section, you will implement a gradient descent algorithm for minimizing the LASSO loss above ($\bar{J}_\lambda(w; D)$),
and you will use a train/validation approach to tune the hyperparameter $\lambda$. Hyperparameter tuning is done following these steps:
* Split the training set into **train/validation** (this does not include the test set).
* For each candidate value of $\lambda$, fit a LASSO model to the training data and compute the **validation loss**, which is the MSE over the validation data (**without** the regularizing term). 
* Choose the $\lambda$ that prodcues the lowest validation loss, and then evaluate the error of the resulting model (using the MSE) on the test set, which was not used in any of these steps.

Note that the regularizing term ($\lambda\left\|w\right\|_1$) is only used when fitting the linear function to the training set.
It is not used in when computing the loss implied by that linear function on the validation or test sets.
The rationale is that the model error is still measured by the MSE, and the regularizing term is just meant to guide the training process to a model with weights close to 0, possibly at the expense of slightly higher error.

## 10. Nearest Neighbors Regression (10 pts)

In K-nearest neighbor regression, we predict the target value of a new point by looking at its **k most similar training samples** (nearest in feature space) and averaging their target values. The most common way to define distance between samples is using Euclidean distance
($d(u,v) = \|u-v\|_2$). Unlike linear regression, which assumes a global linear relationship between features and target, K‑NN regression is a **non‑parametric local method**: predictions are based directly on the training data and can fit complex shapes, but they are typically slower and sensitive to the choice of feature scaling and distance metric.

---
**Complete the function `knn_predict` in `hw1.py`.**
The function implements prediciton based on K-NN regression and uses Euclidean distances to identify the k most similar training samples to the given sample.

---
The code cell below reduced the data to the following **three features**: `sqft_living`, `sqft_lot`, `bedrooms`,
and then invokes the `knn_predict` function to predict labels of all samples in the test set.
It then computes the test MSE of the K-NN predictor and compares it the the MSE of the linear predictor trained using these three features.


```python
from hw1 import knn_predict

# 3 most interesting features (provided in notebook)
selected_features = ['sqft_living', 'sqft_lot', 'bedrooms']
X_train_reduced = data_df[selected_features].iloc[idx_train].values.astype('float64')
X_test_reduced = data_df[selected_features].iloc[idx_test].values.astype('float64')

k = 3
y_pred_knn = np.array(knn_predict(X_train_reduced, y_train, X_test_reduced, k))

knn_mse = np.mean((y_pred_knn - y_test) ** 2)

X_train_reduced_with_bias = add_bias_feature(X_train_reduced)
X_test_reduced_with_bias = add_bias_feature(X_test_reduced)

w_star = compute_pinv(X_train_reduced_with_bias, y_train)
y_pred_linear_regression = X_test_reduced_with_bias @ w_star
lr_mse = np.mean((y_pred_linear_regression - y_test) ** 2)
print('K-NN MSE:', knn_mse)
print('Linear regression MSE:', lr_mse)
```
